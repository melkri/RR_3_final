```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from IPython.display import Markdown as md
from scipy.interpolate import make_interp_spline
df = pd.read_csv('Doctor_Who_Episodes_2005-21.csv')
df['Original_air_date'] = pd.to_datetime(df['Original_air_date'])
```

```python
average_viewers = df['UK_viewers(millions)'].mean()
average_score = df['Appreciation _Index'].mean()
most_beloved_episode = df.loc[df['Appreciation _Index'].idxmax(), 'Episode']
least_liked_episode = df.loc[df['Appreciation _Index'].idxmin(), 'Episode']
most_beloved_doctor = df.groupby('The_Doctor')['Appreciation _Index'].mean().idxmax()
least_beloved_doctor = df.groupby('The_Doctor')['Appreciation _Index'].mean().idxmin()
```

## Average Rating: 84

Data for the whole below analysis can be found on [kaggle](https://www.kaggle.com/datasets/jacklacey/doctor-who-episodes-20052021).

#### Show Statistics:

```python
md(f"""
- Average Viewers: __{average_viewers:.1f} mln__
- Average Score: __{average_score:.0f}__
- Most Beloved Episode: __{most_beloved_episode}__
- Least Liked Episode: __{least_liked_episode}__
- Most Beloved Doctor: __{most_beloved_doctor}__
- Least Beloved Doctor: __{least_beloved_doctor}__
""")
```

#### Viewership over Time

```python
plt.figure(figsize=(10, 6))
ax = df.plot(x='Original_air_date', y='UK_viewers(millions)', color='blue', linewidth=2, alpha=0.7)
ax.set_title('UK Viewers over Time', fontsize=16)
ax.set_xlabel('Original Air Date', fontsize=14)
ax.set_ylabel('UK Viewers (millions)', fontsize=14)

ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.AutoDateFormatter(mdates.AutoDateLocator()))

ax.tick_params(axis='x', rotation=45)
ax.grid(False)
ax.set_ylim(0)
plt.tight_layout()
plt.show()
```

#### Appreciation over Time

```python
plt.figure(figsize=(10, 6))
ax = df.plot(x='Original_air_date', y='Appreciation _Index', color='green', linewidth=2, alpha=0.7)
ax.set_title('Appreciation Index over Time', fontsize=16)
ax.set_xlabel('Original Air Date', fontsize=14)
ax.set_ylabel('Appreciation Index', fontsize=14)

ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.AutoDateFormatter(mdates.AutoDateLocator()))

ax.tick_params(axis='x', rotation=45)
ax.grid(False)
plt.tight_layout()
plt.show()
```

As we can see the overall popularity of the show is decreasing, from around __11 mln__ viewers at 2010 to around __6 mln__ viewers at 2020. Additionally rating tho stays at a relatively high level is also decreasing. Obviously a big part of that could be a raise in popularity of __streaming platforms__ and increase in number of overall productions.
